```python
import pandas as pd
import os
from sklearn.ensemble import RandomForestClassifier
```

# Read the CSV and Perform Basic Data Cleaning

```python
exoplanets = pd.read_csv("Resources/exoplanet_data.csv")
# Drop the null columns where all values are null
exoplanets = exoplanets.dropna(axis='columns', how='all')
# Drop the null rows
exoplanets = exoplanets.dropna()
exoplanets.head()
```

```python
exoplanets.describe()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
target = exoplanets["koi_disposition"]
```

```python
data = exoplanets.drop("koi_disposition", axis=1)
feature_names = data.columns
data.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaler_train = scaler.fit(X_train)
X_scaler_test = scaler.fit(X_test)
X_train_scaled = X_scaler_train.transform(X_train)
X_test_scaled = X_scaler_test.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
model.fit(X_train_scaled, y_train)
```

```python
model.predict(X_test_scaled)
```

# Train the Model



```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
confusion_matrix(y_test, model.predict(X_test_scaled))
```
